# 02 — Phase 1: Data Overview
**Football Players Stats 2024/25**

Quick overview of the dataset — distributions, counts, and basic summaries.

## Setup

```python
import pandas as pd
import plotly.express as px

player_df = pd.read_csv('Dataset\player_clean.csv')
league_df = pd.read_csv('Dataset\league_clean.csv')

print(f"player_df: {player_df.shape}")
print(f"league_df: {league_df.shape}")
```

## 2.1 — Players by League

```python
LEAGUE_COLORS = {
    'Premier League': '#3d195b', 'La Liga': '#ee8707',
    'Serie A': '#008fd7', 'Bundesliga': '#d3010c', 'Ligue 1': '#091c3e',
}

by_league = league_df.groupby('league')['player_name'].nunique().reset_index()
by_league.columns = ['league', 'players']

fig = px.bar(by_league, x='league', y='players', color='league',
             color_discrete_map=LEAGUE_COLORS,
             title='Number of Players by League (2024/25)',
             labels={'players': 'Number of Players', 'league': ''},
             text='players')
fig.update_traces(textposition='outside')
fig.update_layout(height=420, showlegend=False, plot_bgcolor='white',
                  yaxis=dict(showgrid=True, gridcolor='#f0f0f0'))
fig.show()
```

## 2.2 — Players by Position

```python
POSITION_COLORS = {'FW': '#e63946', 'MF': '#2a9d8f', 'DF': '#457b9d', 'GK': '#6a4c93'}

by_pos = player_df['primary_position'].value_counts().reset_index()
by_pos.columns = ['position', 'count']

fig = px.bar(by_pos, x='position', y='count', color='position',
             color_discrete_map=POSITION_COLORS,
             title='Players by Position (2024/25)',
             labels={'count': 'Number of Players', 'position': ''},
             text='count',
             category_orders={'position': ['DF', 'MF', 'FW', 'GK']})
fig.update_traces(textposition='outside')
fig.update_layout(height=420, showlegend=False, plot_bgcolor='white',
                  yaxis=dict(showgrid=True, gridcolor='#f0f0f0'))
fig.show()
```

## 2.3 — Age Distribution

```python
fig = px.histogram(player_df, x='age', nbins=20,
                   color_discrete_sequence=['#3d195b'],
                   title='Player Age Distribution (2024/25)',
                   labels={'age': 'Age', 'count': 'Number of Players'})
fig.update_layout(height=420, plot_bgcolor='white',
                  yaxis=dict(showgrid=True, gridcolor='#f0f0f0'))
fig.show()
print(f"Average age: {player_df['age'].mean():.1f}")
print(f"Youngest: {player_df['age'].min()} | Oldest: {player_df['age'].max()}")
```

## 2.4 — Minutes Played Distribution

```python
fig = px.histogram(player_df, x='minutes_played', nbins=30,
                   color_discrete_sequence=['#00b377'],
                   title='Minutes Played Distribution — 900 min threshold shown',
                   labels={'minutes_played': 'Minutes Played'})
fig.add_vline(x=900, line_dash='dash', line_color='red',
              annotation_text='900 min threshold',
              annotation_position='top right')
fig.update_layout(height=420, plot_bgcolor='white',
                  yaxis=dict(showgrid=True, gridcolor='#f0f0f0'))
fig.show()
print(f"Players with 900+ minutes: {(player_df['minutes_played'] >= 900).sum()}")
print(f"Players below 900 minutes: {(player_df['minutes_played'] < 900).sum()}")
```
