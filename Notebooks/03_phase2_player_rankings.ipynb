# 03 — Phase 2: Player Rankings
**Football Players Stats 2024/25**

Top 10 rankings by position — Forwards, Midfielders, Defenders, Goalkeepers.

> All rankings filtered to players with **900+ minutes** (sufficient_minutes == True)

## Setup

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

player_df = pd.read_csv('Dataset\player_clean.csv')

LEAGUE_COLORS = {
    'Premier League': '#3d195b', 'La Liga': '#ee8707',
    'Serie A': '#008fd7', 'Bundesliga': '#d3010c', 'Ligue 1': '#091c3e',
}

ranked     = player_df[player_df['sufficient_minutes'] == True].copy()
forwards   = ranked[ranked['primary_position'] == 'FW'].copy()
mids       = ranked[ranked['primary_position'] == 'MF'].copy()
defenders  = ranked[ranked['primary_position'] == 'DF'].copy()
keepers    = ranked[ranked['primary_position'] == 'GK'].copy()

print(f"Forwards: {len(forwards)} | Mids: {len(mids)} | Defenders: {len(defenders)} | GKs: {len(keepers)}")

def plot_top10(df, col, title, xlabel, min_filter=None):
    data = df.copy()
    if min_filter:
        filter_col, min_val = min_filter
        data = data[data[filter_col] >= min_val]
    data = data.nlargest(10, col).sort_values(col, ascending=True)
    fig = px.bar(data, x=col, y='player_name', orientation='h',
                 color='league', color_discrete_map=LEAGUE_COLORS,
                 hover_data=['club', 'league', col],
                 title=title, labels={col: xlabel, 'player_name': ''},
                 text=col)
    fig.update_traces(textposition='outside', texttemplate='%{text:.1f}')
    fig.update_layout(height=450, plot_bgcolor='white', legend_title='League',
                      xaxis=dict(showgrid=True, gridcolor='#f0f0f0'),
                      yaxis=dict(showgrid=False),
                      margin=dict(l=10, r=60, t=50, b=40))
    fig.show()
```

## 3.1 — Forwards

```python
plot_top10(forwards, 'goals', 'Top 10 Forwards — Goals (2024/25)', 'Goals')
plot_top10(forwards, 'assists', 'Top 10 Forwards — Assists (2024/25)', 'Assists')
plot_top10(forwards, 'xg', 'Top 10 Forwards — xG (2024/25)', 'xG')
plot_top10(forwards, 'goals_vs_xg', 'Top 10 Forwards — Goals vs xG (Overperformers)', 'Goals minus xG')
plot_top10(forwards, 'goals_per_shot_on_target', 'Top 10 Forwards — Conversion Rate (min 20 SoT)',
           'Goals per Shot on Target', min_filter=('shots_on_target', 20))

# Scatter — xG vs Goals
fig = px.scatter(forwards, x='xg', y='goals', color='league',
                 color_discrete_map=LEAGUE_COLORS, size='minutes_played', size_max=15,
                 hover_data=['player_name', 'club', 'goals', 'xg', 'goals_vs_xg'],
                 title='Forwards — xG vs Actual Goals (2024/25)',
                 labels={'xg': 'Expected Goals (xG)', 'goals': 'Actual Goals'})
max_val = max(forwards['xg'].max(), forwards['goals'].max()) + 2
fig.add_shape(type='line', x0=0, y0=0, x1=max_val, y1=max_val,
              line=dict(color='gray', dash='dash', width=1.5))
fig.update_layout(height=550, plot_bgcolor='white',
                  xaxis=dict(showgrid=True, gridcolor='#f0f0f0'),
                  yaxis=dict(showgrid=True, gridcolor='#f0f0f0'))
fig.show()
```

## 3.2 — Midfielders

```python
plot_top10(mids, 'key_passes', 'Top 10 Midfielders — Key Passes (2024/25)', 'Key Passes')
plot_top10(mids, 'progressive_passes', 'Top 10 Midfielders — Progressive Passes (2024/25)', 'Progressive Passes')
plot_top10(mids, 'shot_creating_actions', 'Top 10 Midfielders — Shot Creating Actions (2024/25)', 'SCA')
plot_top10(mids, 'pass_completion_pct', 'Top 10 Midfielders — Pass Completion % (min 500 passes)',
           'Pass Completion %', min_filter=('passes_attempted', 500))
```

## 3.3 — Defenders

```python
plot_top10(defenders, 'tackles_won', 'Top 10 Defenders — Tackles Won (2024/25)', 'Tackles Won')
plot_top10(defenders, 'interceptions', 'Top 10 Defenders — Interceptions (2024/25)', 'Interceptions')
plot_top10(defenders, 'clearances', 'Top 10 Defenders — Clearances (2024/25)', 'Clearances')
plot_top10(defenders, 'tackles_interceptions', 'Top 10 Defenders — Tackles + Interceptions (2024/25)', 'Tackles + Interceptions')

defenders['aerial_duels_total'] = defenders['aerial_duels_won'] + defenders['aerial_duels_lost']
plot_top10(defenders, 'aerial_duel_win_pct', 'Top 10 Defenders — Aerial Duel Win % (min 50 duels)',
           'Aerial Duel Win %', min_filter=('aerial_duels_total', 50))
```

## 3.4 — Goalkeepers

```python
plot_top10(keepers, 'save_pct', 'Top 10 Goalkeepers — Save % (min 50 shots faced)',
           'Save %', min_filter=('shots_on_target_faced', 50))
plot_top10(keepers, 'post_shot_xg_difference', 'Top 10 Goalkeepers — PSxG Difference (2024/25)',
           'Post Shot xG Difference')
plot_top10(keepers, 'clean_sheets', 'Top 10 Goalkeepers — Clean Sheets (2024/25)', 'Clean Sheets')
```
