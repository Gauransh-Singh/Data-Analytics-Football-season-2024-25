## 1.3 — Handle Null Values

```python
# Drop 8 rows with missing Age (fringe players with <700 mins)
print(f"Rows before: {len(df)}")
df.dropna(subset=['age'], inplace=True)
print(f"Rows after dropping null age: {len(df)}")

# Fill missing nationality with 'Unknown'
df['nationality'] = df['nationality'].fillna('Unknown')
print(f"Nationality nulls remaining: {df['nationality'].isnull().sum()}")
```

## 1.4 — Fill GK Columns with 0 for Outfield Players

```python
gk_cols = [
    'goals_conceded', 'goals_conceded_per90', 'saves',
    'shots_on_target_faced', 'clean_sheets', 'gk_wins',
    'gk_draws', 'gk_losses', 'post_shot_xg',
    'post_shot_xg_difference', 'psxg_difference_per90',
    'penalties_faced', 'penalties_saved',
    'penalties_missed_by_shooter', 'avg_pass_length',
    'crosses_stopped', 'sweeper_actions', 'sweeper_actions_per90',
    'avg_sweeper_distance', 'gk_throws', 'crosses_faced',
    'goal_kicks', 'gk_passes_completed', 'gk_passes_attempted',
    'penalties_scored_against'
]
df[gk_cols] = df[gk_cols].fillna(0)
print("GK columns filled with 0 for outfield players")
```

## 1.5 — Fix Data Types

```python
# Convert float to int where appropriate
int_cols = [
    'age', 'birth_year', 'goals_conceded', 'saves',
    'shots_on_target_faced', 'clean_sheets', 'gk_wins',
    'gk_draws', 'gk_losses', 'penalties_faced', 'penalties_saved',
    'penalties_missed_by_shooter', 'penalties_scored_against',
    'gk_passes_completed', 'gk_passes_attempted', 'gk_throws',
    'goal_kicks', 'crosses_stopped', 'crosses_faced', 'sweeper_actions'
]
for col in int_cols:
    df[col] = df[col].astype(int)

print("Data types fixed")
print(df[['age', 'birth_year']].dtypes)
```

## 1.6 — Clean League Names

```python
cleanup_map = {
    'eng Premier League': 'Premier League',
    'es La Liga':         'La Liga',
    'fr Ligue 1':         'Ligue 1',
    'it Serie A':         'Serie A',
    'de Bundesliga':      'Bundesliga'
}
df['league'] = df['league'].replace(cleanup_map)
print("Leagues after cleaning:", df['league'].unique())
```

## 1.7 — Create Primary Position Column

```python
df['primary_position'] = df['position'].apply(lambda x: x.split(',')[0])
print("Primary position distribution:")
print(df['primary_position'].value_counts())
```

## 1.8 — Create league_df and player_df

```python
# league_df — one row per player per club stint (for league-level analysis)
league_df = df.copy()
print(f"league_df shape: {league_df.shape}")
```

```python
# Check multi-club players
multi_club = df[df.duplicated(subset='player_name', keep=False)]
print(f"Players with multiple club rows: {multi_club['player_name'].nunique()}")
print(multi_club[['player_name', 'club', 'league', 'minutes_played', 'goals', 'assists']].head(10))
```

```python
# Sort by player name and minutes descending
# so the club with most minutes comes first in each group
df = df.sort_values(by=['player_name', 'minutes_played'], ascending=[True, False])

# Define column groups for aggregation
identity_cols = ['player_name', 'nationality', 'position', 'club', 'league', 'age', 'birth_year', 'primary_position']

counting_cols = [
    'matches_played', 'matches_started', 'minutes_played', 'games_equivalent',
    'goals', 'assists', 'goals_assists', 'nonpenalty_goals', 'penalty_goals',
    'penalty_attempts', 'yellow_cards', 'red_cards', 'progressive_carries',
    'progressive_passes', 'progressive_receptions', 'xg', 'npxg', 'xag',
    'npxg_plus_xag', 'goals_assists_nopenalty', 'xg_plus_xag', 'xa',
    'goals_vs_xg', 'npgoals_vs_xg', 'assists_vs_xag', 'shots',
    'shots_on_target', 'freekick_shots', 'passes_completed', 'passes_attempted',
    'pass_total_distance', 'pass_progressive_distance', 'key_passes',
    'passes_final_third', 'passes_penalty_area', 'crosses_penalty_area',
    'live_ball_passes', 'dead_ball_passes', 'freekick_passes', 'through_balls',
    'switches', 'crosses', 'throw_ins', 'corner_kicks', 'inswing_corners',
    'outswing_corners', 'straight_corners', 'tackles', 'tackles_won',
    'tackles_def_third', 'tackles_mid_third', 'tackles_att_third',
    'dribbles_challenged', 'dribbles_challenged_lost', 'blocks', 'shots_blocked',
    'passes_blocked', 'interceptions', 'tackles_interceptions', 'clearances',
    'errors_leading_to_shot', 'shot_creating_actions', 'sca_live_pass',
    'sca_dead_pass', 'sca_takeons', 'sca_shots', 'sca_fouls_drawn',
    'sca_defensive_actions', 'goal_creating_actions', 'touches',
    'touches_own_box', 'touches_def_third', 'touches_mid_third',
    'touches_att_third', 'touches_opp_box', 'live_ball_touches',
    'dribbles_attempted', 'dribbles_successful', 'times_tackled_dribbling',
    'carries', 'carry_progressive_distance', 'carries_final_third',
    'carries_penalty_area', 'miscontrols', 'dispossessed', 'passes_received',
    'second_yellow_cards', 'fouls_committed', 'fouls_drawn', 'offsides',
    'penalties_won', 'penalties_conceded', 'own_goals', 'recoveries',
    'aerial_duels_won', 'aerial_duels_lost', 'goals_conceded',
    'shots_on_target_faced', 'saves', 'gk_wins', 'gk_draws', 'gk_losses',
    'clean_sheets', 'penalties_faced', 'penalties_scored_against',
    'penalties_saved', 'penalties_missed_by_shooter', 'post_shot_xg',
    'post_shot_xg_difference', 'gk_passes_completed', 'gk_passes_attempted',
    'goal_kicks', 'gk_throws', 'crosses_faced', 'crosses_stopped',
    'sweeper_actions',
]

rate_cols = [
    'shot_accuracy_pct', 'goals_per_shot', 'goals_per_shot_on_target',
    'avg_shot_distance', 'npxg_per_shot', 'pass_completion_pct',
    'tackle_success_pct', 'shot_creating_actions_per90',
    'goal_creating_actions_per90', 'dribble_success_pct',
    'tackled_while_dribbling_pct', 'aerial_duel_win_pct',
    'goals_conceded_per90', 'save_pct', 'clean_sheet_pct',
    'post_shot_xg_per_shot', 'psxg_difference_per90', 'gk_pass_completion_pct',
    'long_pass_pct', 'avg_pass_length', 'cross_stop_pct',
    'sweeper_actions_per90', 'avg_sweeper_distance',
]

# Aggregate counting cols
identity_df  = df.groupby('player_name')[identity_cols[1:]].first()
counting_df  = df.groupby('player_name')[counting_cols].sum()

# Weighted average for rate cols
def weighted_avg(group):
    result = {}
    for col in rate_cols:
        valid = group[['minutes_played', col]].dropna()
        if valid.empty or valid['minutes_played'].sum() == 0:
            result[col] = None
        else:
            result[col] = (valid[col] * valid['minutes_played']).sum() / valid['minutes_played'].sum()
    return pd.Series(result)

rate_df = df.groupby('player_name').apply(weighted_avg)

# Merge all three
player_df = identity_df.join(counting_df).join(rate_df).reset_index()
print(f"player_df shape: {player_df.shape}")
```

## 1.9 — Add Features & Save

```python
# Sufficient minutes flag (900 mins = ~10 full games)
player_df['sufficient_minutes'] = player_df['minutes_played'] >= 900

# Per-90 columns
per90_cols = ['goals', 'assists', 'shots', 'shots_on_target',
              'key_passes', 'progressive_carries', 'progressive_passes',
              'tackles', 'interceptions', 'clearances', 'xg', 'xag']

for col in per90_cols:
    player_df[f'{col}_per90'] = (player_df[col] / player_df['games_equivalent']).round(2)

# Add sufficient_minutes to league_df too
league_df['sufficient_minutes'] = league_df['minutes_played'] >= 900
league_df['primary_position']   = league_df['position'].apply(lambda x: x.split(',')[0])

print(f"Sufficient minutes: {player_df['sufficient_minutes'].value_counts().to_dict()}")
print(f"\nSaving cleaned files...")
player_df.to_csv('player_clean.csv', index=False)
league_df.to_csv('league_clean.csv', index=False)
print("✅ Saved: player_clean.csv and league_clean.csv")
```
